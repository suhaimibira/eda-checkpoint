```python
import pandas as pd 
import numpy as np 

features_df = pd.read_csv('C:/Users/Sabria/Desktop/Python/Database for GA/Features data set.csv')
sales_df = pd.read_csv('C:/Users/Sabria/Desktop/Python/Database for GA/sales data-set.csv', header=0)
stores_df = pd.read_csv('C:/Users/Sabria/Desktop/Python/Database for GA/stores data-set.csv')

print(features_df.index.duplicated().sum())
print(sales_df.index.duplicated().sum())
print(stores_df.index.duplicated().sum())
sales_df.columns = sales_df.columns.str.strip()
print(sales_df.columns)

sales_df['Date'] = pd.to_datetime(sales_df['Date'], errors= 'coerce')

```

## Get to Know the Numbers

```python
features_df.reset_index(drop=True, inplace=True)
sales_df.reset_index(drop=True, inplace=True)
stores_df.reset_index(drop=True, inplace=True)

combined_df = combined_df.reset_index()

combined_df = pd.concat([features_df, sales_df, stores_df], axis=1)
combined_df.reset_index(drop=True, inplace=True)

first_10 = combined_df.head(10)
# shows the first 10 columns
last_10 = combined_df.tail(10)
# shows the last 10 columns

unique_values = combined_df.nunique()
# shows the count of unique values

info = combined_df.info()
# provides the data type of each column, the total # of entries and number of non-null values for each column

description = combined_df.describe(include='all')
# provides the count, mean, standard deviation, and min, max and quartiles 

columns_index = combined_df.columns
# shows all the columns. 

print(info)
print(description)
print(columns_index)
print(first_10)
print(last_10)
print(unique_values)

print(features_df.index.duplicated().sum())
print(sales_df.index.duplicated().sum())
print(stores_df.index.duplicated().sum())
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# need to see correlation between if it is a holiday and weekly sales
# correlation between weekly_sales and temperature and fuel price 
# see the correlation between CPI and weekly sales. 

import pandas as pd
import matplotlib.pyplot as plt

print('Fig.1')

plt.figure(figsize=(10,6))
colors = (sales_df['IsHoliday']).map({True: 'red', False: 'blue'})

plt.scatter(x=combined_df['Temperature'], y=combined_df['Weekly_Sales'], c=colors, label ='Holiday status', alpha=0.7)

plt.xlabel('Temperature(F)')
plt.ylabel('Weekly Sales($)')
plt.title('Weekly Sales vs Temperature(Colored by Holiday Status)')

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Holiday', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Non-Holiday', markerfacecolor='blue', markersize=10)
]
plt.legend(handles=legend_elements)
plt.show()

print('Fig. 2') 

plt.figure(figsize=(10,6))
colors = (sales_df['IsHoliday']).map({True: 'red', False: 'blue'})

plt.scatter(x=combined_df['Fuel_Price'], y=combined_df['Weekly_Sales'], c=colors, label ='Holiday status', alpha=0.7)

plt.xlabel('Fuel Price($)')
plt.ylabel('Weekly Sales($)')
plt.title('Weekly Sales vs Fuel Price (Colored by Holiday Status)')

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Holiday', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Non-Holiday', markerfacecolor='blue', markersize=10)
]
plt.legend(handles=legend_elements)
plt.show()

print('Fig. 3') 

plt.figure(figsize=(10,6))
colors = (sales_df['IsHoliday']).map({True: 'red', False: 'blue'})

plt.scatter(x=combined_df['Unemployment'], y=combined_df['Weekly_Sales'], c=colors, label ='Holiday status', alpha=0.7)

plt.xlabel('Unemployment Rate(%)')
plt.ylabel('Weekly Sales($)')
plt.title('Weekly Sales vs Unemployment Rate (Colored by Holiday Status)')

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Holiday', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Non-Holiday', markerfacecolor='blue', markersize=10)
]
plt.legend(handles=legend_elements)
plt.show()

print('Fig. 4') 

plt.figure(figsize=(10,6))
colors = (sales_df['IsHoliday']).map({True: 'red', False: 'blue'})

plt.scatter(x=combined_df['CPI'], y=combined_df['Weekly_Sales'], c=colors, label ='Holiday status', alpha=0.7)

plt.xlabel('CPI')
plt.ylabel('Weekly Sales($)')
plt.title('Weekly Sales vs Consumer Price Index (Colored by Holiday Status)')

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Holiday', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Non-Holiday', markerfacecolor='blue', markersize=10)
]
plt.legend(handles=legend_elements)
plt.show()
```
